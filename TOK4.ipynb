```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Функции для метода Якоби
def x1_jacobi(x2, x3):
    return (5+4*x2+x3)/7

def x2_jacobi(x1, x3):
    return (12-x3)/3

def x3_jacobi(x1, x2):
    return (13-5*x2)/6
```

```python
# Функция для метода Гаусса-Зейделя
def gauss_seidel_iteration(x1, x2, x3):
    x1_new = (5+4*x2+x3)/7
    x2_new = (12-x3)/3
    x3_new = (13-5*x2_new)/6
    return x1_new, x2_new, x3_new
```

```python
# Начальные значения переменных
x1_j, x2_j, x3_j = 0, 0, 0
x1_gs, x2_gs, x3_gs = 0, 0, 0
# Количество итераций
iterations = 10
# Списки для хранения значений на каждой итерации
jacobi_x1, jacobi_x2, jacobi_x3 = [], [], []
gs_x1, gs_x2, gs_x3 = [], [], []
# Порог ошибки
tolerance = 0.001
```

```python
# Итерационный процесс для метода Якоби и метода Гаусса-Зейделя
iterations = 0

while True:
    # Запоминаем текущие значения для графиков
    jacobi_x1.append(x1_j)
    jacobi_x2.append(x2_j)
    jacobi_x3.append(x3_j)

    gs_x1.append(x1_gs)
    gs_x2.append(x2_gs)
    gs_x3.append(x3_gs)

    # Обновляем значения методом Якоби (старые значения используются)
    x1_j_new = (5 + 4 * x2_j + x3_j) / 7
    x2_j_new = (12 - x3_j) / 3
    x3_j_new = (13 - 5 * x2_j) / 6

    # Обновляем значения методом Гаусса-Зейделя (используем новые значения сразу)
    x1_gs_new, x2_gs_new, x3_gs_new = gauss_seidel_iteration(x1_gs, x2_gs, x3_gs)

    # Вычисляем максимальную ошибку
    error_jacobi = max(abs(x1_j - x1_j_new), abs(x2_j - x2_j_new), abs(x3_j - x3_j_new))
    error_gs = max(abs(x1_gs - x1_gs_new), abs(x2_gs - x2_gs_new), abs(x3_gs - x3_gs_new))

    # Обновляем значения
    x1_j, x2_j, x3_j = x1_j_new, x2_j_new, x3_j_new
    x1_gs, x2_gs, x3_gs = x1_gs_new, x2_gs_new, x3_gs_new

    iterations += 1

    # Проверка условия выхода
    if error_jacobi < tolerance and error_gs < tolerance:
        break
```

```python
# Добавляем последние значения после итераций
jacobi_x1.append(x1_j)
jacobi_x2.append(x2_j)
jacobi_x3.append(x3_j)

gs_x1.append(x1_gs)
gs_x2.append(x2_gs)
gs_x3.append(x3_gs)
```

```python
print(jacobi_x1)
```

```python
print(gs_x1)
```

```python
plt.figure(figsize=(10, 6))

# График x1
plt.plot(range(iterations + 1), jacobi_x1, marker='o', linestyle='--', label="Якоби: x1")
plt.plot(range(iterations + 1), gs_x1, marker='s', linestyle='-', label="Гаусс-Зейдель: x1")

plt.xlabel("Количество итераций")
plt.ylabel("Значение переменной")
plt.title("Сравнение методов Якоби и Гаусса-Зейделя")
plt.legend()
plt.grid(True)

plt.show()
```
