```python
import math
import matplotlib.pyplot as plt

# Исходная функция
def f(x):
    return 6 * math.exp(-0.2 * x) * math.sin(2 * x + 0.3)

# Аналитическая первая производная:
# f'(x) = 6*exp(-0.2*x)*(2*cos(2*x+0.3) - 0.2*sin(2*x+0.3))
def f_prime(x):
    return 6 * math.exp(-0.2 * x) * (2 * math.cos(2 * x + 0.3) - 0.2 * math.sin(2 * x + 0.3))

# Аналитическая вторая производная:
# f''(x) = 6*exp(-0.2*x)*(-0.8*cos(2*x+0.3) - 3.96*sin(2*x+0.3))
def f_double_prime(x):
    return 6 * math.exp(-0.2 * x) * (-0.8 * math.cos(2 * x + 0.3) - 3.96 * math.sin(2 * x + 0.3))

# Функция для генерации диапазона значений
def frange(start, stop, step):
    x = start
    while x <= stop + 1e-9:  # чтобы включить верхнюю границу
        yield x
        x += step
```

```python
h = 0.03
# Задаём узлы на отрезке [-0.05, 0.55]
x_vals = [round(i, 3) for i in frange(-0.05, 0.55, h)]
y_vals = [f(x) for x in x_vals]
```

```python
# -------------------------------
# Аппроксимации первой производной
# -------------------------------

# 1) Прямая разностная формула (forward difference)
def fwd_diff(x, h):
    return (f(x + h) - f(x)) / h

# 2) Задняя разностная формула (backward difference)
def bwd_diff(x, h):
    return (f(x) - f(x - h)) / h

# 3) Центральная разностная формула
def cen_diff(x, h):
    return (f(x + h) - f(x - h)) / (2 * h)

# 4) 4-точечная центральная формула:
#    f'(x) ~ (-f(x+2h) + 8f(x+h) - 8f(x-h) + f(x-2h))/(12h)
def four_point_diff(x, h):
    return (-f(x + 2 * h) + 8 * f(x + h) - 8 * f(x - h) + f(x - 2 * h)) / (12 * h)
```

```python
# Для каждой формулы вычисляем аппроксимацию только в тех узлах, где имеются все необходимые точки.
# Для forward: x <= max - h
x_fwd = [x for x in x_vals if x <= 0.55 - h]
y_fwd = [fwd_diff(x, h) for x in x_fwd]

# Для backward: x >= min + h
x_bwd = [x for x in x_vals if x >= -0.05 + h]
y_bwd = [bwd_diff(x, h) for x in x_bwd]

# Для центральной: x в [min+h, max-h]
x_cen = [x for x in x_vals if x >= -0.05 + h and x <= 0.55 - h]
y_cen = [cen_diff(x, h) for x in x_cen]

# Для 4-точечной: x в [min+2h, max-2h]
x_four = [x for x in x_vals if x >= -0.05 + 2 * h and x <= 0.55 - 2 * h]
y_four = [four_point_diff(x, h) for x in x_four]

# Аналитическое значение первой производной в соответствующих узлах
y_fwd_anal = [f_prime(x) for x in x_fwd]
y_bwd_anal = [f_prime(x) for x in x_bwd]
y_cen_anal = [f_prime(x) for x in x_cen]
y_four_anal = [f_prime(x) for x in x_four]
```

```python
# Расчёт абсолютных ошибок для первой производной
err_fwd = [abs(approx - true) for approx, true in zip(y_fwd, y_fwd_anal)]
err_bwd = [abs(approx - true) for approx, true in zip(y_bwd, y_bwd_anal)]
err_cen = [abs(approx - true) for approx, true in zip(y_cen, y_cen_anal)]
err_four = [abs(approx - true) for approx, true in zip(y_four, y_four_anal)]
```

```python
# Построение графика первой производной (аналитическая vs аппроксимированные)
plt.figure(figsize=(10, 6))
plt.plot(x_fwd, y_fwd_anal, 'ko', label="Аналитическая f'(x)")
plt.plot(x_fwd, y_fwd, 'g-', label="Прямая разн. формула")
plt.plot(x_bwd, y_bwd, 'b', label="Задняя разн. формула")
plt.plot(x_cen, y_cen, 'r-', label="Центральная разн. формула")
plt.plot(x_four, y_four, 'mo', label="4-точечная формула")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.title("Первая производная: аналитическая и приближённые значения")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Построение графика ошибок для первой производной
plt.figure(figsize=(10, 6))
plt.plot(x_fwd, err_fwd, 'go-', label="Ошибка прямой разн. формулы")
plt.plot(x_bwd, err_bwd, 'bo-', label="Ошибка задней разн. формулы")
plt.xlabel("x")
plt.ylabel("Абсолютная ошибка")
plt.title("Погрешности аппроксимации первой производной")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(x_cen, err_cen, 'ro-', label="Ошибка центральной разн. формулы")
plt.plot(x_four, err_four, 'mo-', label="Ошибка 4-точечной формулы")
plt.xlabel("x")
plt.ylabel("Абсолютная ошибка")
plt.title("Погрешности аппроксимации первой производной")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# -------------------------------
# Аппроксимации второй производной
# -------------------------------

# 1) Прямая разностная формула для второй производной:
#    f''(x) ~ (f(x+2h) - 2*f(x+h) + f(x))/(h**2)
def fwd_diff2(x, h):
    return (f(x + 2 * h) - 2 * f(x + h) + f(x)) / (h ** 2)

# 2) Задняя разностная формула для второй производной:
#    f''(x) ~ (f(x) - 2*f(x-h) + f(x-2h))/(h**2)
def bwd_diff2(x, h):
    return (f(x) - 2 * f(x - h) + f(x - 2 * h)) / (h ** 2)

# 3) Центральная разностная формула для второй производной:
#    f''(x) ~ (f(x+h) - 2*f(x) + f(x-h))/(h**2)
def cen_diff2(x, h):
    return (f(x + h) - 2 * f(x) + f(x - h)) / (h ** 2)

# 4) 4-точечная центральная формула для второй производной:
#    f''(x) ~ (-f(x+2h) + 16*f(x+h) - 30*f(x) + 16*f(x-h) - f(x-2h))/(12h**2)
def four_point_diff2(x, h):
    return (-f(x + 2 * h) + 16 * f(x + h) - 30 * f(x) + 16 * f(x - h) - f(x - 2 * h)) / (12 * h ** 2)
```

```python
# Определяем диапазоны, где доступны все необходимые точки.
# Для forward: x <= max - 2h
x_fwd2 = [x for x in x_vals if x <= 0.55 - 2 * h]
y_fwd2 = [fwd_diff2(x, h) for x in x_fwd2]

# Для backward: x >= min + 2h
x_bwd2 = [x for x in x_vals if x >= -0.05 + 2 * h]
y_bwd2 = [bwd_diff2(x, h) for x in x_bwd2]

# Для центральной: x в [min+h, max-h]
x_cen2 = [x for x in x_vals if x >= -0.05 + h and x <= 0.55 - h]
y_cen2 = [cen_diff2(x, h) for x in x_cen2]

# Для 4-точечной: x в [min+2h, max-2h]
x_four2 = [x for x in x_vals if x >= -0.05 + 2 * h and x <= 0.55 - 2 * h]
y_four2 = [four_point_diff2(x, h) for x in x_four2]

# Аналитическое значение второй производной
y_fwd2_anal = [f_double_prime(x) for x in x_fwd2]
y_bwd2_anal = [f_double_prime(x) for x in x_bwd2]
y_cen2_anal = [f_double_prime(x) for x in x_cen2]
y_four2_anal = [f_double_prime(x) for x in x_four2]
```

```python
# Расчёт ошибок для второй производной
err_fwd2 = [abs(approx - true) for approx, true in zip(y_fwd2, y_fwd2_anal)]
err_bwd2 = [abs(approx - true) for approx, true in zip(y_bwd2, y_bwd2_anal)]
err_cen2 = [abs(approx - true) for approx, true in zip(y_cen2, y_cen2_anal)]
err_four2 = [abs(approx - true) for approx, true in zip(y_four2, y_four2_anal)]
```

```python
# График второй производной (аналитическая vs аппроксимированные)
plt.figure(figsize=(10, 6))
plt.plot(x_fwd2, y_fwd2_anal, 'k-', label="Аналитическая f''(x)")
plt.plot(x_fwd2, y_fwd2, 'g-', label="Прямая разн. формула 2-ой")
plt.plot(x_bwd2, y_bwd2, 'b-', label="Задняя разн. формула 2-ой")
plt.plot(x_cen2, y_cen2, 'r-', label="Центральная разн. формула 2-ой")
plt.plot(x_four2, y_four2, 'mo', label="4-точечная формула 2-ой")
plt.xlabel("x")
plt.ylabel("f''(x)")
plt.title("Вторая производная: аналитическая и приближённые значения")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# График ошибок для второй производной
plt.figure(figsize=(10, 6))
plt.plot(x_fwd2, err_fwd2, 'go-', label="Ошибка прямой формулы 2-ой")
plt.plot(x_bwd2, err_bwd2, 'bo-', label="Ошибка задней формулы 2-ой")
plt.xlabel("x")
plt.ylabel("Абсолютная ошибка")
plt.title("Погрешности аппроксимации второй производной")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(x_cen2, err_cen2, 'ro-', label="Ошибка центральной формулы 2-ой")
plt.plot(x_four2, err_four2, 'mo-', label="Ошибка 4-точечной формулы 2-ой")
plt.xlabel("x")
plt.ylabel("Абсолютная ошибка")
plt.title("Погрешности аппроксимации второй производной")
plt.legend()
plt.grid(True)
plt.show()
```
