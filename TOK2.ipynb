```python
import math
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 6*math.exp(-0.2*x)*math.sin(2*x+0.3)
```

```python
def frange(x, y, jump):
    while x < y:
        yield x
        x += jump
```

```python
x = [round(i,3) for i in frange(-0.05, 0.55, 0.03)]
y = [round(f(i),3) for i in x]
h = 0.03
```

```python
plt.plot(x, y)
```

```python
def f_pro_straight(x, h):
    return (f(x+h) - f(x))/h
x0 = [round(i,3) for i in frange(-0.05, 0.55, 0.03)]
y0 = [round(f_pro_straight(i, h),3) for i in x]

def f_pro_center(x, h):
    return (f(x+h) - f(x-h))/(2*h)
x1 = [round(i,3) for i in frange(-0.05, 0.55, 0.03)]
y1 = [round(f_pro_center(i, h),3) for i in x]

def f_pro_back(x, h):
    return (f(x) - f(x-h))/h
x2 = [round(i,3) for i in frange(-0.05, 0.55, 0.03)]
y2 = [round(f_pro_back(i, h),3) for i in x]

plt.plot(x, y)
plt.plot(x0, y0, color="green")
plt.plot(x1, y1, color="red")
plt.plot(x2, y2, color="blue")
plt.show()
```

```python
def f_2pro_straight(x, h):
    return (f(x+(2*h)) - 2*f(x+h) + f(x))/(h**2)
x0 = [round(i,3) for i in frange(-0.05, 0.55, 0.03)]
y0 = [round(f_2pro_straight(i, h),3) for i in x]

def f_2pro_center(x, h):
    return (f(x+h) - 2*f(x) + f(x-h))/(h**2)
x1 = [round(i,3) for i in frange(-0.05, 0.55, 0.03)]
y1 = [round(f_2pro_center(i, h),3) for i in x]

def f_2pro_back(x, h):
    return (f(x) - 2*f(x-h) + f(x-2*h))/(h**2)
x2 = [round(i,3) for i in frange(-0.05, 0.55, 0.03)]
y2 = [round(f_2pro_back(i, h),3) for i in x]

plt.plot(x, y)
plt.plot(x0, y0, color="green")
plt.plot(x1, y1, color="red")
plt.plot(x2, y2, color="blue")
plt.show()
```
